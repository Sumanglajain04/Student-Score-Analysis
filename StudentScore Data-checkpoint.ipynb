# Drop unnamed column

```python
df = df.drop('Unnamed: 0', axis = 1)
print(df.head())
```

# change weekly study hours column

```python
df["WklyStudyHours"] = df["WklyStudyHours"].str.replace("10-May", "5-10")
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#from the above chart we have analysed that:
#the number of females in the data is more than the number of males
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})

gb = gb.rename(columns=lambda x: x.replace('ore', ''))
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot = True)
plt.show()
```

```python
#from the above chat we have concluded that the education of the parents have a good impact on their score
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1 = gb1.rename(columns=lambda x: x.replace('ore', ''))
print(gb1)
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb1,annot = True)
plt.title("Relationship between Parent's Marital Status and Student's Score")
plt.show()
```

```python
#from the above chart we have concluded that there is no/negligible
#impact on the students's score to their parent's marotal status
```

# Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
l = ["group A" , "group B" , "group C" , "group E"]

mlist = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"]]
print(mlist)
plt.title("Distribution of Groups")
plt.pie(mlist, labels=l, autopct = "%1.2f%%")
plt.show()
```
